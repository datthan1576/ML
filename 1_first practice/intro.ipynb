### Giải quyết bài toán phân loại nhị phân 

### Mô hình học
Mô hình thuật toán $\mathfrak{F}$ trong học máy là một hợp các hàm mà nó hoạt động từ tập đối tượng và tập nhãn,cụ thể trong bài toán này:
$$\mathfrak{F} = \{f| f: \mathbb{R}^n \to \{-1, 1\}, \text{còn một số giới hạn nào đó}\},$$
Thông thường $\mathfrak{F}$ là họ tham số của các hàm, có nghĩa là $f$ khác nhau bới một số tham số. Ví dụ đơn giản cho một họ tham số cho bài toán phân loại nhị phân là họ của bộ phân loại tuyến tính:
$$\mathfrak{F}_{bcl} = \left\{f\bigr(\theta, \mathbf{x}\bigr)=\text{sign}\bigr(\theta^{\mathsf{T}}\mathbf{x}\bigr)\bigr| \theta \in \mathbb{R}^{n} \right\}.$$


### Hàm mất mát 
Học máy luôn giải quyết bài toán lựa chọn hàm từ tập $\mathfrak{F}$. Để chọn được hàm này cần một số tiêu chí,theo những tiêu chí này các hàm sẽ được xếp hạng từ xấu nhất cho tới tốt nhất. Để có thể làm được việc này chúng ta tạo nên một chức năng  $\mathcal{L}$, nó sẽ đưa ra cho mỗi $f \in \mathfrak{F}$ một số tương ứng từ $\mathbb{R}_+$. Trong học máy, chức năng chất lượng thường được định nghĩa là một số lỗi trong mẫu. Chức năng chất lượng có thể được biểu diễn như sau:
$$\mathcal{L}\bigr(f, \mathbf{X}, \mathbf{y}\bigr) = \sum_{i=1}^l\mathcal{q}\bigr(f, \mathbf{x}_i, y_i\bigr),$$
với $q$ là một hàm lỗi nào đó trên đối tượng $\mathbf{x}$. Hàm chức năng $\mathcal{L}$ gọi là rủi ro thực nghiệm (empirical risk).

### Bài toán tối ưu hóa
Sau đó chúng ta cần đưa ra nhiệm vụ tối ưu hóa để chọn ra $f \in \mathfrak{F}$. Để làm được điều này chúng ta phải tối thiếu hóa rủi ro thực nghiệm:
$$\hat{f} = \arg \min_{f \in \mathfrak{F}} \mathcal{L}\bigr(f, \mathbf{X}, \mathbf{y}\bigr).$$

Quan trọng! Kết quả hàm $\hat{f}$ phụ thuộc vào mẫu $\left(\mathbf{X}, \mathbf{y}\right)$, có nghĩa là với những mẫu dữ liệu khác nhau thì hàm tối ưu sẽ khác nhau.

Trong trường hợp của chúng tôi, hàm chất lượng sẽ có dạng sau:
$$\mathcal{L}\bigr(\theta, \mathbf{X}, \mathbf{y}\bigr) = \sum_{i=1}^l\bigr[f\bigr(\theta, \mathbf{x}_i\bigr) \not= y_i\bigr],$$
và vấn đề tối ưu hóa được viết lại thành:
$$\hat{\theta} = \arg \min_{\theta \in \mathbb{R}^n} \sum_{i=1}^l\bigr[f\bigr(\theta, \mathbf{x}_i\bigr) \not= y_i\bigr].$$
Và trên thực tế, trong tương lai, chúng ta sẽ giải quyết vô số vấn đề như vậy, để tìm ra tham số tối ưu. Lời giải của bài toán phân loại nhị phân tuyến tính sẽ có trong bài giảng tiếp theo. Bây giờ chúng tôi đang sử dụng các thư viện để giải quyết vấn đề này. Hơn nữa trong ví dụ, tham số sẽ được tìm thấy $\hat{\theta}$ không phải là một giải pháp trực tiếp cho vấn đề tối ưu hóa này, nhưng được sửa đổi một chút, nhưng nhiều hơn về vấn đề đó ở phần sau của bài giảng tiếp theo.

### Tìm vectơ tối ưu của các tham số

Xem xét các ma trận
1. Ma trận đối tượng $\mathbf{X} \in \mathbb{R}^{l\times (n+1)}$
2. Ma trận nhãn $\mathbf{y} \in \{-1,1\}^l$

Lưu ý rằng chúng ta đã nhúng các đối tượng vào một không gian có chiều cao hơn bằng cách thêm một đặc trưng, đặc trưng này đối với tất cả các đối tượng sẽ bằng $1$.

```python
# Ma trận đối tượng
X = binary_dataset.iloc[:, binary_dataset.columns != 'Class'].values
# Vectơ các nhãn - Cần được reshape để có thể đưa vào mô hình học
y = binary_dataset.iloc[:, binary_dataset.columns == 'Class'].values.reshape(-1)
# Chồng thêm 1 đặc trưng cho các đối tượng
X = np.array(np.hstack([X, np.ones([len(X), 1])]), dtype=np.float64)
# Chuyển đổi kiểu dữ liệu của nhãn
y = np.array(y, dtype=np.int64)
```

```python
# Tạo model và train
model = LogisticRegression(random_state=0, max_iter=2000)
_ = model.fit(X, y)
```

Chúng ta nhận được tham số tối ưu cho $\hat{\theta}$
